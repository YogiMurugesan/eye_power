```python
def Add(x):
  if (x>=2) & (x<=2.25):
    return "Normal"
  else:
    return "abnormal"
```

```python
x=pd.get_dummies(x)
```

```python
x.head()
```

```python
m=x[['Sphere','Cylinder','Axis','Age']]
y=x['out']
```

```python
from sklearn.model_selection import train_test_split
m_train,m_test,y_train,y_test=train_test_split(m,y,test_size=0.3,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(max_depth=6,random_state=0,n_estimators=10)
model.fit(m_train,y_train)
```

```python
y_pred=model.predict(m_test)
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,y_pred)**(0.4)
mse
```
